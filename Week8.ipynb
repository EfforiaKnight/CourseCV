```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('IMG_7826.jpg')
img = cv2.resize(img, (0, 0), fx=0.15, fy=0.15, interpolation=cv2.INTER_AREA)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray_crop = img_gray[30:100,30:100].copy()
hist = cv2.calcHist([img_gray_crop], [0], None, [256], [0, 256])
hist = cv2.normalize(hist, hist)
```

```python
img_gray_blur_3 = cv2.blur(img_gray,(3,3))
img_gray_blur_5 = cv2.blur(img_gray,(5,5))
img_gray_blur_10 = cv2.blur(img_gray,(10,10))
img_gray_blur_50 = cv2.blur(img_gray,(50,50))
img_gray_blur_100 = cv2.blur(img_gray,(100,100))
```

```python
# show the images
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(20, 20))
ax[0].set_axis_off()
ax[0].set_title('img_gray_blur_3', fontsize=16, color="red")
ax[0].imshow(img_gray_blur_3, cmap='gray')

ax[1].set_axis_off()
ax[1].set_title('img_gray_blur_5', fontsize=16, color="red")
ax[1].imshow(img_gray_blur_5, cmap='gray')

ax[2].set_axis_off()
ax[2].set_title('img_gray_blur_10', fontsize=16, color="red")
ax[2].imshow(img_gray_blur_10, cmap='gray')

ax[3].set_axis_off()
ax[3].set_title('img_gray_blur_50', fontsize=16, color="red")
ax[3].imshow(img_gray_blur_50, cmap='gray')

ax[4].set_axis_off()
ax[4].set_title('img_gray_blur_100', fontsize=16, color="red")
ax[4].imshow(img_gray_blur_100, cmap='gray')

plt.show()
```

```python
# show the images
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))
ax1[0].set_axis_off()
ax1[0].set_title('Img', fontsize=16, color="red")
ax1[0].imshow(img_gray, cmap='gray')

ax1[1].set_axis_off()

ax2[0].set_axis_off()
ax2[0].set_title('Img Crop', fontsize=16, color="red")
ax2[0].imshow(img_gray_crop, cmap='gray')

ax2[1].set_title('Hist', fontsize=16, color="red")
ax2[1].plot(hist)

plt.show()
```

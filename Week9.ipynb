```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('IMG_7826.jpg')
img = cv2.resize(img, (0, 0), fx=0.15, fy=0.15, interpolation=cv2.INTER_AREA)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
median_5 = cv2.medianBlur(img_gray, 5)
median_7 = cv2.medianBlur(img_gray, 7)
median_15 = cv2.medianBlur(img_gray, 15)
median_31 = cv2.medianBlur(img_gray, 31)
```

```python
# show the images
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(30, 30))
ax[0].set_axis_off()
ax[0].set_title('img_gray', fontsize=16, color="red")
ax[0].imshow(img_gray, cmap='gray')

ax[1].set_axis_off()
ax[1].set_title('median_5', fontsize=16, color="red")
ax[1].imshow(median_5, cmap='gray')

ax[2].set_axis_off()
ax[2].set_title('median_7', fontsize=16, color="red")
ax[2].imshow(median_7, cmap='gray')

ax[3].set_axis_off()
ax[3].set_title('median_15', fontsize=16, color="red")
ax[3].imshow(median_15, cmap='gray')

ax[4].set_axis_off()
ax[4].set_title('median_31', fontsize=16, color="red")
ax[4].imshow(median_31, cmap='gray')

plt.show()
```

### Perform image denoising using adaptive-filtering-local-noise-filter https://angeljohnsy.blogspot.com/2011/12/adaptive-filtering-local-noise-filter.html

```python
def ataptive_mean_filter(img, M=5, N=5):
    w, h = img.shape
    
    lmean = np.zeros_like(img)
    lvar = np.zeros_like(img)
    
    for i in range(w-M):
        for j in range(h-N):
            temp = img[i:i+M, j:j+N]
            lmean[i, j] = np.mean(temp)
            lvar[i, j] = np.var(temp)
    
    return 
```

```python
new_img = ataptive_mean_filter(img_gray)
```
